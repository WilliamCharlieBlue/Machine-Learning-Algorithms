# 13.7 Stacking

## 13.7.1 Voting Classifier
- Voting Classifier：对多个预测结果进行投票的结果作为最终结果

## 1.7.2 Stacking
- Stacking：将多个预测结果作为输入，再训练一个新的模型，用新模型的结果作为最终结果
    - 如果有N层作为训练，则需要将训练数据分为N+1份，有一份是用于最终模型的训练
    - 非常类似于神经网络，层数增多会产生深度学习的模型
    - Scikit-learn中没有提供Stacking的接口
