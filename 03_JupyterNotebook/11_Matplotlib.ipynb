### 11.1 plot
- 折/曲线图 横轴是特征，纵轴是label
    - plt.plot(x, siny, label="sin(x)")  绘制一条曲线，曲线名称sin(x)
    - plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")  绘制第二条曲线，红色，虚线，曲线名称cos(x)
    - plt.axis([-1, 11, -2, 2])  x轴取值范围[-1,11], y轴取值范围[-2, 2]
    - plt.xlabel("x axis")  x轴名称
    - plt.ylabel("y value") y轴名称
    - plt.legend()  将x曲线和y曲线的图例显示出来
    - plt.title("Welcom to my ML learning class")  图的主标题
    - plt.show()  将整幅图显示出来

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
# 从1-10，切分100段
x = np.linspace(0, 10, 100)
```

```python
x
x.size
```

```python
y = np.sin(x)
```

```python
y
y.shape
```

```python
plt.plot(x,y)
# 先执行一系列操作，然后用show()函数展示出来
plt.show()
```

```python
siny = y.copy()
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
# color="red" "blue" "green" '#eeefff' '0.75'
plt.plot(x, cosy, color='red')
plt.show()
```

```python
plt.plot(x, siny)
# linestyle=":" "-." "--" "-"
plt.plot(x, cosy, color="red", linestyle="--")
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2])
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.axis([-1, 11, -2, 2])
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.axis([-1, 11, -2, 2])
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcom to my ML learning class")
plt.show()
```

### 11.2 Scatter Plot
- 散点图 横纵坐标都是特征，label用颜色来表示
    - x = np.random.normal(0,1,10000)  10000个正太分布的点
    - y = np.random.normal(0,1,10000)  10000个正太分布的点
    - plt.scatter(x, y, alpha=0.1)  散点都打在图上，不透明度为0.1
    - plt.show()  把整个图显示出来

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```

```python
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
# alpha不透明度0-1
plt.scatter(x, y, alpha=0.1)
plt.show()
```
