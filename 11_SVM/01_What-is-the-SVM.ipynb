# 11.1 支持向量机 SVM (Support Vector Machine)

## 11.1.1 SVM想解决的问题
- 模型的未来泛化能力的考量直接放在算法内部，而不是预处理或者模型正则化步骤
- SVM尝试寻找一个最优的决策边界，距离两个类别的最近的样本最远
- 即最大化margin（最大化2d）
    - 线性可分问题 -> Hard margin SVM
    - 非线性可分问题 -> Soft margin SVM



## 11.1.2 SVM既能解决分类问题，也能解决回归问题
- 分类和回归的思想是相反的
    - 分类问题：两个支撑向量之间尽量不要有点
    - 回归问题：两个支撑向量之间的点尽可能多
