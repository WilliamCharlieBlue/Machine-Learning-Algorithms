## 11.6.3 高斯核(FBR核)
- 将每一个样本点映射到一个无穷维的特征空间
- 对于每一个数据点都是landmark
- (m, n)的数据映射成了(m, m)的数据
- 计算开销非常大，适用于初始数据维度非常高，但样本数量较少，例如自然语言处理
