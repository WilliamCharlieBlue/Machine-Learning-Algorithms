# 8.10 L1正则、L2正则、弹性网

## 8.10.1 Lp范数和Ln正则
    - 明可夫斯基距离衍生出来 Lp范数
    - 而Lp范数从而得到Ln正则项



## 8.10.2 L1正则项和L2正则项
    - Lasso对应L1正则项
    - Ridge对应L2正则项


## 8.10.3 L0正则
   - 实际用L1取代，因为L0正则的优化是一个NP难的问题
    - L0是离散自由化的问题，可能要穷举所有使的θ等于0的组合，再决定哪些θ为0。
    - L0不能使用梯度下降法以及具体的数据公式找出最优解
    - L1里可以得到大量θ=0的情况，使得可以作为特征选择



## 8.10.4 弹性网
    - 引入了一个新的超参数r
    - 结合了L1和L2两种优势


## 8.10.5 Ridge、LASSO、Elastic Net的选择
    - Ridge 计算精准，特征较多时计算量巨大
    - LASSO 具有特征选择功能，但容易出错
    - Elastic Net 结合了Ridge和LASSO的优点
    - 实际使用：先尝试Ridge，再尝试Elastic Net
