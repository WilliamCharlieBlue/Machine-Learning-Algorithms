# 8.7 偏差方差权衡

## 8.7.1 模型误差
### 模型误差 = 偏差(Bias) + 方差(Variance) + 不可避免的误差


## 8.7.2 偏差
    - 对问题的假设是不正确的
    - 例如：非线性数据使用线性回归
    - 在特征工程层面 数据特征的提取是主要原因
    - 在具体算法层面 欠拟合underfitting是主要原因

## 8.7.3 方差
    - 数据的一点点扰动都会较大地影响模型。
    - 如高阶多项式回归
    - 通常地原因是模型太过复杂，学习了过多地噪音
    - 在具体算法层面 欠拟合overfitting是主要原因

## 8.7.4 偏差和方差的矛盾
    - 天生高方差的算法
        - 非参数学习通常都是高方差的算法，因为不对数据进行任何假设
        - 如kNN
    - 天生高偏差的算法
        - 参数学习通常都是高偏差的算法，因为对数据具有极强的假设
        - 如线性回归
    - 偏差和方差的调整
        - 如 kNN中的k
            - k越小，模型越复杂，方差越大，偏差越小
        - 如 线性回归中使用的多项式回归的degree
            - degree越小，模型越简单，偏差越大，方差越小
    - 偏差和方差的矛与盾
        - 通常降低偏差，会提高方差；降低方差，会提高方差

## 8.7.5 偏差和方差的挑战
    - 问题层面
        - 疾病、金融等问题的理解还是很肤浅，对问题的假设并不准确
        - 特征工程的挑战来自偏差
    - 算法层面
        - 机器学习的主要挑战来自方差
        - 解决overfitting的问题，提高泛化能力是主要的
        - 降低方差的几个主要手段：
            1. 降低模型复杂度
            2. 减少数据维度；降噪
            3. 增加样本数（神经网络/深度学习的一个基本要求）
            4. 使用验证集
            5. 模型正则化
