# 8.1 多项式回归

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3,3,size=100)
X = x.reshape(-1, 1)
```

```python
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(x,y)
plt.show()
```

## 8.1.1 使用sklearn的LinearRegression来拟合

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, y_predict, color='r')
plt.show()
```

## 8.1.2 增加一个x^2 特征，即使用多项式回归

```python
(X**2).shape
```

```python
X2 = np.hstack([X, X**2])
```

```python
X2.shape
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
y_predict2 = lin_reg2.predict(X2)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color='r')
plt.show()
```

```python
# y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
# 生成数据的时候x系数为1，x^2的系数为0.5
lin_reg2.coef_
```
