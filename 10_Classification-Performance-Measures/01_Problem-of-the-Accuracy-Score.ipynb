# 10.1 准确度的陷阱和混淆矩阵

## 10.1.1 准确度陷阱

- 如果癌症产生的概率只有0.1%
    - 我们的系统预测所有人都是健康，即可达到99.9%的准确率


- 如果癌症产生的概率只有0.01%
    - 我们的系统预测所有人都是健康，即可达到99.99%的准确率

## 10.1.2 极度偏斜数据（Skewed Data）
- 极度偏斜的数据只使用分类准确度是远远不够的
- 使用混淆矩阵进一步分析

## 10.1.3 混淆矩阵（Confusion Matrix）
- 对于二分类问题
    - 行代表真实值，列代表预测值
    - 预测为negative则为N，预测为positive则为P
    - 如果预测的正确则为T，如果预测的错误则为F
